```python
from selenium import webdriver
from selenium.webdriver.common.action_chains import ActionChains
import time
#from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By

#driver_path="/Users/apple/Desktop/driver/chromedriver"
driver = webdriver.Firefox()
driver.get("https://www.python.org/")

action = ActionChains(driver)

#action.click()
#action.move()


# tab = driver.find_element_by_class_name('donate-button') for chrome browser

#For firefox browser
tab = driver.find_element(By.CLASS_NAME, 'donate-button')

print(tab)

print("*********************************** END ************************************************")

element1 = driver.find_element(By.ID,'mainnav')

print(element1)

print("*********************************** END ************************************************")


element2 = driver.find_elements(By.ID,'mainnav')

print(element2)

print("*********************************** END ************************************************")

element3 = driver.find_element(By.TAG_NAME,'p')

print(element3)

print("*********************************** END ************************************************")


element4 = driver.find_elements(By.TAG_NAME,'p')

print(element4)

print("*********************************** END ************************************************")


element5 = driver.find_element(By.LINK_TEXT,'Donate') # Just the text found on the web page

print(element5)

print("*********************************** END ************************************************")


element6 = driver.find_elements(By.NAME,'q') # CValue of attribute name="q"

print(element6)

print("*********************************** END ************************************************")

print("At this point, i want to check just if click action will work")
action.click(on_element=tab)

print("*********************************** END ************************************************")

# Performing Click Actions

# action.click(on_element=tab) - Performs a click action on the element
# action.double_click(on_element=tab) - Double clicks the specified element
# action.click_and_hold(on_element=tab) - Clicks and holds an element;
# It is always good to perform a release action after using the click_and_hold method so that you
# can release the hold on the element
# action.click_and_hold(on_element=tab)
# action.release(tab)


action.perform()
time.sleep(10) # Delays the shutting down of the browser window

driver.quit()
```
